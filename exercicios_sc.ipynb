```python
import random
alfabeto = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
```

Questão 1

```python
def cifrar_cesar(mensagem, chave):
  global alfabeto
  mensagem = mensagem.upper()
  mensagem_cifrada = ""

  for i in mensagem:
    if i in alfabeto:
      mensagem_cifrada += alfabeto[(alfabeto.index(i) + chave) % 26]
    else:
      mensagem_cifrada += i
  return mensagem_cifrada
```

```python
def decifrar_cesar(mensagem, chave):
  global alfabeto
  mensagem = mensagem.upper()
  mensagem_decifrada = ""

  for i in mensagem:
    if i in alfabeto:
      mensagem_decifrada += alfabeto[(alfabeto.index(i) - chave) % 26]
    else:
      mensagem_decifrada += i
  return mensagem_decifrada
```

```python
def tenta_todos(mensagem):
  global alfabeto
  mensagem = mensagem.upper()
  palavras_comuns = ["O", "A", "E", "OS", "AS", "DE", "DO", "QUE", "COM", "SE", "EM", "NOS"]
  mensagem_decifrada = ""
  melhor_resultado = ""
  melhor_numero_tentativas = 0

  for i in range(26):
    for j in mensagem:
      if j in alfabeto:
        mensagem_decifrada += alfabeto[(alfabeto.index(j) - i) % 26]
      else:
        mensagem_decifrada += j

    tentativas = sum(1 for palavra in palavras_comuns if palavra in mensagem_decifrada.upper())
    if tentativas > melhor_numero_tentativas:
      melhor_numero_tentativas = tentativas
      melhor_resultado = mensagem_decifrada

    print(f"Tentativa {i + 1}: {mensagem_decifrada}")
    mensagem_decifrada = ""
  return melhor_resultado
```

```python
def analise_frequencia(mensagem):
  global alfabeto
  mensagem = mensagem.upper()
  frequencias = { "A": 14.63, "B": 1.04, "C": 3.88, "D": 4.99, "E": 12.57, "F": 1.02, "G": 1.30,
                  "H": 1.28, "I": 6.18, "J": 0.40, "K": 0.02, "L": 2.78, "M": 4.74, "N": 5.05,
                  "O": 10.73, "P": 2.52, "Q": 1.20, "R": 6.53, "S": 7.81, "T": 4.34, "U": 4.63,
                  "V": 1.67, "W": 0.01, "X": 0.21, "Y": 0.01, "Z": 0.47 }
  frequencias = sorted(frequencias, key=frequencias.get, reverse=True)
  frequencias_cifradas = {}
  mensagem_decifrada = ""
  cont = sum(1 for i in mensagem if i in alfabeto)

  for i in mensagem:
    if i in alfabeto:
      if i not in frequencias_cifradas:
        frequencias_cifradas[i] = 1
      else:
        frequencias_cifradas[i] += 1

  for j in frequencias_cifradas.keys():
    frequencias_cifradas[j] = (frequencias_cifradas[j] / cont) * 100

  frequencias_cifradas = sorted(frequencias_cifradas, key=frequencias_cifradas.get, reverse=True)
  provavel_chave = (alfabeto.index(frequencias_cifradas[0]) - alfabeto.index(frequencias[0])) % 26
  return provavel_chave
```

```python
def cifrar_vigenere(mensagem, chave):
  global alfabeto
  mensagem = mensagem.upper()
  mensagem_cifrada = ""
  indice_chave = 0

  for i in range(len(mensagem)):
    if mensagem[i] in alfabeto:
      mensagem_cifrada += alfabeto[(alfabeto.index(mensagem[i]) + alfabeto.index(chave[indice_chave % len(chave)])) % 26]
      indice_chave += 1
    else:
      mensagem_cifrada += mensagem[i]

  return mensagem_cifrada
```

```python
def decifrar_vigenere(mensagem, chave):
  global alfabeto
  mensagem = mensagem.upper()
  mensagem_decifrada = ""
  indice_chave = 0

  for i in range(len(mensagem)):
    if mensagem[i] in alfabeto:
      mensagem_decifrada += alfabeto[(alfabeto.index(mensagem[i]) - alfabeto.index(chave[indice_chave % len(chave)]))]
      indice_chave += 1
    else:
      mensagem_decifrada += mensagem[i]

  return mensagem_decifrada
```

```python
def gerar_chave():
  alfabeto = list("ABCDEFGHIJKLMNOPQRSTUVWXYZ")
  random.shuffle(alfabeto)

  return "".join(alfabeto)
```

```python
def cifra_permutacao_aleatoria(mensagem, chave):
  global alfabeto
  mensagem = mensagem.upper()
  mensagem_cifrada = ""

  for i in range(len(mensagem)):
    if mensagem[i] in alfabeto:
      mensagem_cifrada += chave[alfabeto.index(mensagem[i])]
    else:
      mensagem_cifrada += mensagem[i]

  return mensagem_cifrada
```

```python
def decifra_permutacao_aleatoria(mensagem, chave):
  global alfabeto
  mmensagem = mensagem.upper()
  mensagem_decifrada = ""

  for i in range(len(mensagem)):
    if mensagem[i] in alfabeto:
      mensagem_decifrada += alfabeto[chave.index(mensagem[i])]
    else:
      mensagem_decifrada += mensagem[i]

  return mensagem_decifrada
```

```python
def cifra_transposicao(mensagem, chave):
  global alfabeto
  mensagem = mensagem.replace(" ", "").upper()
  mensagem_cifrada = ""
  ordem = sorted(range(len(chave)), key=lambda k: chave[k])
  matriz = [list(mensagem[i:i+len(chave)]) for i in range(0, len(mensagem), len(chave))]

  while len(matriz[-1]) < len(chave):
    matriz[-1].append(alfabeto[random.randint(0, 25)])

  for i in ordem:
      for j in matriz:
          mensagem_cifrada += j[i]

  return mensagem_cifrada
```

```python
mensagem = "AO VERME QUE PRIMEIRO ROEU AS FRIAS CARNES DO MEU CADAVER DEDICO COMO SAUDOSA LEMBRANCA ESTAS MEMORIAS POSTUMAS"
chave_cesar = 5
chave_vigenere = "MACHADODEASSIS"
chave_transposicao = "LEGADO"

print("------------------------------------------------------------------")
print("1.")
mensagem_cifrada_cesar = cifrar_cesar(mensagem, chave_cesar)
print(f"Mensagem Cifrada: {mensagem_cifrada_cesar}")

mensagem_decifrada_cesar = decifrar_cesar(mensagem_cifrada_cesar, chave_cesar)
print(f"Mensagem Decifrada: {mensagem_decifrada_cesar}")

print("------------------------------------------------------------------")
print("2.")
mensagem_forcaBruta = tenta_todos(mensagem_cifrada_cesar)
print(f"Força Bruta: {mensagem_forcaBruta}")

print("------------------------------------------------------------------")
print("3.")
provavel_chave = analise_frequencia(mensagem_cifrada_cesar)
print(f"Chave Por Análise de Frequência: {provavel_chave}")

print("------------------------------------------------------------------")
print("4.")
mensagem_cifrada_vigenere = cifrar_vigenere(mensagem, chave_vigenere)
mensagem_decifrada_vigenere = decifrar_vigenere(mensagem_cifrada_vigenere, chave_vigenere)
print(f"Mensagem Cifrada: {mensagem_cifrada_vigenere}")
print(f"Mensagem Decifrada: {mensagem_decifrada_vigenere}")

print("------------------------------------------------------------------")
print("5.")
chave_aleatoria = gerar_chave()
mensagem_cifrada_permutacao = cifra_permutacao_aleatoria(mensagem, chave_aleatoria)
mensagem_decifrada_permutacao = decifra_permutacao_aleatoria(mensagem_cifrada_permutacao, chave_aleatoria)
print(f"Chave Aleatória: {chave_aleatoria}")
print(f"Mensagem Cifrada: {mensagem_cifrada_permutacao}")
print(f"Mensagem Decifrada: {mensagem_decifrada_permutacao}")

print("------------------------------------------------------------------")
print("6.")
mensagem_cifrada_transposicao = cifra_transposicao(mensagem, chave_transposicao)
print(f"Mensagem Cifrada: {mensagem_cifrada_transposicao}")
```
