```python
def cifrar_cesar(mensagem, chave):
  alfabeto = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
  mensagem_cifrada = ""

  for i in mensagem:
    if i.upper() in alfabeto:
      mensagem_cifrada += alfabeto[(alfabeto.index(i.upper()) + chave) % 26]
    else:
      mensagem_cifrada += i
  return mensagem_cifrada
```

```python
def decifrar_cesar(mensagem, chave):
  alfabeto = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
  mensagem_decifrada = ""

  for i in mensagem:
    if i.upper() in alfabeto:
      mensagem_decifrada += alfabeto[(alfabeto.index(i.upper()) - chave) % 26]
    else:
      mensagem_decifrada += i
  return mensagem_decifrada
```

```python
def tenta_todos(mensagem):
  alfabeto = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
  palavras_comuns = ["O", "A", "E", "OS", "AS", "DE", "DO", "QUE", "COM", "SE", "EM", "NOS"]
  mensagem_decifrada = ""
  melhor_resultado = ""
  melhor_numero_tentativas = 0

  for i in range(26):
    for j in mensagem:
      if j.upper() in alfabeto:
        mensagem_decifrada += alfabeto[(alfabeto.index(j.upper()) - i) % 26]
      else:
        mensagem_decifrada += j

    tentativas = sum(1 for palavra in palavras_comuns if palavra in mensagem_decifrada.upper())
    if tentativas > melhor_numero_tentativas:
      melhor_numero_tentativas = tentativas
      melhor_resultado = mensagem_decifrada

    print(f"Tentativa {i + 1}: {mensagem_decifrada}")
    mensagem_decifrada = ""
  return melhor_resultado
```

```python
def analise_frequencia(mensagem):
  frequencias = { "A": 14.63, "B": 1.04, "C": 3.88, "D": 4.99, "E": 12.57, "F": 1.02, "G":	1.30,
                  "H": 1.28, "I":	6.18, "J": 0.40, "K": 0.02, "L": 2.78, "M": 4.74, "N": 5.05,
                  "O": 10.73, "P": 2.52, "Q": 1.20, "R": 6.53, "S": 7.81, "T": 4.34, "U": 4.63,
 	  	            "V": 1.67, "W": 0.01, "X": 0.21, "Y": 0.01, "Z": 0.47 }
  frequencias = sorted(frequencias, key=frequencias.get, reverse=True)
  frequencias_cifradas = {}
  alfabeto = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
  mensagem_decifrada = ""
  cont = len(mensagem.replace(" ", ""))

  for i in mensagem:
    if i != " ":
      if i not in frequencias_cifradas.keys():
        frequencias_cifradas[i] = 1
      else:
        frequencias_cifradas[i] += 1

  for j in frequencias_cifradas.keys():
    frequencias_cifradas[j] = (frequencias_cifradas[j] / cont) * 100

  frequencias_cifradas = sorted(frequencias_cifradas, key=frequencias_cifradas.get, reverse=True)
  provavel_chave = (alfabeto.index(frequencias_cifradas[0]) - alfabeto.index(frequencias[0])) % 26
  return provavel_chave
```

```python
mensagem = "AO VERME QUE PRIMEIRO ROEU AS FRIAS CARNES DO MEU CADAVER DEDICO COMO SAUDOSA LEMBRANCA ESTAS MEMORIAS POSTUMAS"
chave = 5

mensagem_cifrada = cifrar_cesar(mensagem, chave)
print(f"Mensagem Cifrada: {mensagem_cifrada}")

print("----------------------------")
mensagem_decifrada = decifrar_cesar(mensagem_cifrada, chave)
print(f"Mensagem Decifrada: {mensagem_decifrada}")

print("----------------------------")
mensagem_forcaBruta = tenta_todos(mensagem_cifrada)
print(f"Força Bruta: {mensagem_forcaBruta}")

print("----------------------------")
provavel_chave = analise_frequencia(mensagem_cifrada)
print(f"Chave Por Análise de Frequência: {provavel_chave}")
```
